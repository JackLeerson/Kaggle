## Missing Data

```python
# Missing and NaN data
missingInfoTrain = pd.DataFrame({'Sum': trainData.isnull().sum(), 'Relative': trainData.isnull().sum()/len(trainData)})
missingInfoTest = pd.DataFrame({'Sum': testData.isnull().sum(), 'Relative': testData.isnull().sum()/len(testData)})
display(missingInfoTrain)
display(missingInfoTest)
```

- Cabin shows a significant quantity of missing data (77%)
- Age is also missing a lot of data (20%)
- Very small number of embarked is missing (<1%)

## Distributions

```python
# Plot distributions of each variable
undesiredColumns = ['Name']
desiredColumns = list(set(trainData.columns) - set(undesiredColumns))
cols = 2
rows = int(math.ceil(len(desiredColumns)/cols))
fig, axs = plt.subplots(nrows=rows, ncols=cols, figsize=(20, 40))
for i, column in enumerate(desiredColumns):
    ax = axs[i//cols][i%cols]
    ax.set_title(column)
    if is_numeric_dtype(trainData[column]):
        trainData[column].plot.hist(bins=12, ax=ax)
    else:
        trainData[column].value_counts().plot.bar(ax=ax)
```

```python
# Plot distributions of each variable
undesiredColumns = ['Name']
desiredColumns = list(set(testData.columns) - set(undesiredColumns))
cols = 2
rows = int(math.ceil(len(desiredColumns)/cols))
fig, axs = plt.subplots(nrows=rows, ncols=cols, figsize=(20, 40))
for i, column in enumerate(desiredColumns):
    ax = axs[i//cols][i%cols]
    ax.set_title(column)
    if is_numeric_dtype(testData[column]):
        testData[column].plot.hist(bins=12, ax=ax)
    else:
        testData[column].value_counts().plot.bar(ax=ax)
```
